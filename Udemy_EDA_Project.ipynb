```python
# ======================================= #
# --------- Univariate Analysis --------- #
# ======================================= #
```

```python
# Let's Check No of Courses which are Paid and Free.

count_courses = data['is_paid'].value_counts()

labels = ['Paid Courses', 'Free Courses']
colors = ['#6A0DAD', '#B497D6']  # dark purple, light purple


plt.figure(figsize=(6, 6))

plt.pie(count_courses,
        labels=labels,
        colors=colors,
        autopct='%1.1f%%',   # show percentage
        startangle=140,      # rotate for better view
        explode=(0.05, 0)   # slightly highlight first slice
        )

plt.title("Free vs Paid Courses Distribution", fontsize=14, fontweight='bold')

plt.axis('equal')
plt.show()
```

#### We are easily understand that Paid Courses are much more than Free Courses. Between the ratio of each course Free Courses are not even a 10% of Paid Courses.

```python
from matplotlib.patches import Shadow
# Let's Check Which subject has No. of Free/ Paid Courses

plt.figure(figsize=(10, 5))

custom_colors = ['#B497D6', '#6A0DAD']
sns.countplot(data=data, x='subject', hue='is_paid', palette=custom_colors)

plt.title("Subject-wise Free vs Paid Courses", fontsize=14, fontweight='bold')
plt.xlabel("Subject", fontsize=9)
plt.ylabel("Number of Courses", fontsize=9)

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

#### As we already know that paid courses are much higher then the free courses. So now We can compare with the subjects, Business Finance has the highest Paid Courses and a lil bite of lesser Web Development has the second highest paid Course. Beside Both of them are also has the highest free courses too.

```python
# Let's Check Which subject has High No. of Subscribers

sub_course=data.groupby('subject')['num_subscribers'].sum().reset_index()
plt.figure(figsize=(10, 5))

custom_colors = ['#B497D6', '#6A0DAD']
sns.barplot(data=sub_course, x='subject', y='num_subscribers', palette=custom_colors)

plt.title("Subject That has More Subscribers", fontsize=14, fontweight='bold')
plt.xlabel("Subject", fontsize=9)
plt.ylabel("Number of Subscribers", fontsize=9)

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

#### We can easily understand that Web Development has the highest subscribers as compare to Business Finance (High Paid Courses). Other side Business Finance Has Highest Paid Courses but has very less No. of subscribers.

```python
# Let's Check Which subject has High Revenue

sum_course=data.groupby('subject')['price'].sum().reset_index()
plt.figure(figsize=(10, 5))

custom_colors = ['#B497D6', '#6A0DAD']
sns.barplot(data=sum_course, x='subject', y='price', palette=custom_colors)

plt.title("Subject That Generate High Revenue", fontsize=14, fontweight='bold')
plt.xlabel("Subject", fontsize=9)
plt.ylabel("Number of Courses", fontsize=9)

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

#### As we already know that Business Finance and Web Development are the two subjects which are highest Paid Courses, But in Generating Revenue There are Web Development Course Which is Highest Revenue Generated Course and on Second Place There is Business Finance Course.

```python
# Let's Check Subject Trend by Year and Price

def plot_subject_sub_trend(data, subject_name):

    df = data[data['subject'] == subject_name]

    yr_trend = df.groupby('Year')['num_subscribers'].sum().reset_index()

    plt.figure(figsize=(10,5))
    sns.set_style("whitegrid")

    sns.lineplot(data=yr_trend,
                 x='Year',
                 y='num_subscribers',
                 marker='o',
                 color="#6A0DAD")

    plt.title(f"{subject_name} Trend by Year and No. of Subscribers", fontsize=14, fontweight='bold')
    plt.xlabel("Year")
    plt.ylabel("Total No. of Subscribers")

    sns.despine()
    plt.tight_layout()
    plt.show()
```

```python
plot_subject_sub_trend(data, 'Business Finance')
```

```python
plot_subject_sub_trend(data, 'Web Development')
```

```python
plot_subject_sub_trend(data, 'Graphic Design')
```

```python
plot_subject_sub_trend(data, 'Musical Instruments')
```

#### No. of Subscribers are changing in each years in every subjects. Business Finance Subject has the good uptrend in starting year until 2015 but after 2015 until 2017 there is a harsh downtrend in number of subscribers. Also other subject graphs are exactly looks same with good uptrend but after 2015 each subject have harsh downtrend in it. In Musical Instruments Subject what ever they gain from 2012, they loose in double in just 2 years.

```python
# Let's Check Subject Trend by Year and Price

def plot_subject_trend(data, subject_name):

    df = data[data['subject'] == subject_name]

    yr_trend = df.groupby('Year')['price'].sum().reset_index()

    plt.figure(figsize=(10,5))
    sns.set_style("whitegrid")

    sns.lineplot(data=yr_trend,
                 x='Year',
                 y='price',
                 marker='o',
                 color="#6A0DAD")

    plt.title(f"{subject_name} Trend by Year and Price", fontsize=14, fontweight='bold')
    plt.xlabel("Year")
    plt.ylabel("Total Price")

    sns.despine()
    plt.tight_layout()
    plt.show()
```

```python
plot_subject_trend(data, 'Business Finance')
```

```python
plot_subject_trend(data, 'Web Development')
```

```python
plot_subject_trend(data, 'Graphic Design')
```

```python
plot_subject_trend(data, 'Musical Instruments')
```

#### Every subject trend looks Really good from 2012, In 2016 there are top revenue in each subject also until that every subject chart looks so positive uptrend. But after 2016 in each subject having downtrend. Moreover Web Development & Music Instruments subjects are in downtrend about nearly 50% from the top. Business Finance is about to down around 25% and For Graphic Design subject It is sustain so good at top price range.

```python
# Let's Check which Course Level has No. of Free/ Paid Courses

courses_level=data.groupby(['course_level','is_paid']).size().reset_index(name="count")

plt.figure(figsize=(10,5))
sns.set_style("whitegrid")

colors=['#B497D6', '#6A0DAD']

sns.barplot(data=courses_level,
            x='course_level',
            y='count',
            hue='is_paid',
            palette=colors)

plt.title("Course_Level That has More Courses", fontsize=14, fontweight='bold')
plt.xlabel("Course Level", fontsize=9)
plt.ylabel("Number of Courses", fontsize=9)

plt.legend(['Free_Courses','Paid_Courses'], title="Course Type")

plt.xticks(rotation=0)
plt.tight_layout()

sns.despine()
plt.show()
```

#### As we can esasily understand that beginner level of courses are have more Free and Paid Courses.

```python
# Let's Check Which Course ahs the no of lectures

rev_data = data.groupby(['subject', 'course_level'])['num_lectures'].sum().reset_index()


plt.figure(figsize=(10,5))
sns.set_style("whitegrid")

palette = sns.color_palette("Purples", len(rev_data['course_level'].unique()))

ax = sns.barplot(data=rev_data,
                 x='subject',
                 y='num_lectures',
                 hue='course_level',
                 palette=palette)

plt.title("Subject-wise Course Level Revenue", fontsize=14, fontweight='bold')
plt.xlabel("Subject")
plt.ylabel("Total Lectures")

plt.xticks(rotation=0)

plt.tight_layout()
sns.despine()
plt.show()
```

#### Web Development has the highest No of Lectures in each Category of Courses, and on The Second position the is a Business Finance Courses has the second highest lectures in each columns.

```python
# Let's Check which courrse level has More No. of Subscribers

rev_data = data.groupby(['subject', 'course_level'])['num_subscribers'].sum().reset_index()


plt.figure(figsize=(10,5))
sns.set_style("whitegrid")

palette = sns.color_palette("Purples", len(rev_data['course_level'].unique()))

ax = sns.barplot(data=rev_data,
                 x='subject',
                 y='num_subscribers',
                 hue='course_level',
                 palette=palette)

plt.title("Subject-wise Course Level Subscribers", fontsize=14, fontweight='bold')
plt.xlabel("Subject")
plt.ylabel("Total Subscribers")

plt.xticks(rotation=0)

plt.tight_layout()
sns.despine()
plt.show()
```

#### Where No of lecture are also hgihest in Web Developments and Here Course Subscribers are also highest in web development and on the second place same as lecture Business Finance course has more subscribers.

```python
# Let's Check which courrse level has More Revenue Generates

rev_data = data.groupby(['subject', 'course_level'])['price'].sum().reset_index()


plt.figure(figsize=(10,5))
sns.set_style("whitegrid")

palette = sns.color_palette("Purples", len(rev_data['course_level'].unique()))

ax = sns.barplot(data=rev_data,
                 x='subject',
                 y='price',
                 hue='course_level',
                 palette=palette)

plt.title("Subject-wise Course Level Revenue", fontsize=14, fontweight='bold')
plt.xlabel("Subject")
plt.ylabel("Total Revenue")

plt.xticks(rotation=0)

plt.tight_layout()
sns.despine()
plt.show()
```

#### The way Web Development has high lecture, high subscribers and now they are also high generated course too. and same as on second place Business Finance has the good enough revenue generate.
